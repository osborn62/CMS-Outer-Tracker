### Plotting
This is a quick overview of plotting using matplotlib and useful plots.
Histograms are very useful plots. They display the distribution of some varaible x.

```python
# Imports
import matplotlib.pyplot as plt # Plotting library
import numpy as np # Useful science tools
from scipy.stats import norm # Useful statistics tools
```

```python
mu = 0 # Set mean of distribution to 0
sigma = 1 # Set standard deviation of distribution to 0
data = np.random.normal(loc = mu, scale = sigma, size = 1000) # Create normal random array of size 1000
plt.hist(data, bins = 'auto') # Plot histogram
# Make the plot labels
plt.xlabel('Value')
plt.ylabel('Counts')
plt.title('Example Histogram!')
plt.show()
```

Okay, now let's plot normalize the histogram and plot a gaussian distribution on top.

```python
x = np.linspace(-4,4,1000) # Create a list from -4 to 4 with 1000 elements in the list
p = norm.pdf(x, mu, sigma) # Create the pdf using the mu and sigma from above
plt.hist(data, bins = 'auto', density = True, alpha = .8) # Density = True normalizes the histogram
plt.plot(x,p, color = 'black')
plt.title('Histogram with Overlapping Gaussian')
plt.ylabel('Proportion')
plt.xlabel('Deviation from Mean')
plt.show()
```

Okay now let's look at another type of plot you will use: 2d histogram. <br>
This is the same idea of a histogram (showing a distribution) but the distribution is in x and y directions. The counts are visualized by a colorscale.

```python
mu = 0
sigma = 1
data_x = np.random.normal(loc = mu, scale = sigma, size = 1000)
data_y = np.random.normal(loc = mu, scale = sigma, size = 1000)
plt.hist2d(data_x,data_y, bins = 20)
plt.colorbar()
plt.xlabel('x Value')
plt.ylabel('y Value')
plt.title('2D Histogram')
plt.show()
```
